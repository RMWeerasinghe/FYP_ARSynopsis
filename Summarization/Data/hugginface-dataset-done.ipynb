## FINDSum-Liquidity

```python
import pandas as pd


liq_test0 = pd.read_csv("/kaggle/input/findsum-text-data/text/FINDSum-Liquidity/liquidity_input_2000/test_liquidity_segment_0_input_2_1000.csv")
liq_test1 = pd.read_csv("/kaggle/input/findsum-text-data/text/FINDSum-Liquidity/liquidity_input_2000/test_liquidity_segment_1_input_2_1000.csv")
liq_test2 = pd.read_csv("/kaggle/input/findsum-text-data/text/FINDSum-Liquidity/liquidity_input_2000/test_liquidity_segment_2_input_2_1000.csv")

liq_train0 = pd.read_csv("/kaggle/input/findsum-text-data/text/FINDSum-Liquidity/liquidity_input_2000/train_liquidity_segment_0_input_2_1000.csv")
liq_train1 = pd.read_csv("/kaggle/input/findsum-text-data/text/FINDSum-Liquidity/liquidity_input_2000/train_liquidity_segment_1_input_2_1000.csv")
liq_train2 = pd.read_csv("/kaggle/input/findsum-text-data/text/FINDSum-Liquidity/liquidity_input_2000/train_liquidity_segment_2_input_2_1000.csv")

liq_val0 = pd.read_csv("/kaggle/input/findsum-text-data/text/FINDSum-Liquidity/liquidity_input_2000/val_liquidity_segment_0_input_2_1000.csv")
liq_val1 = pd.read_csv("/kaggle/input/findsum-text-data/text/FINDSum-Liquidity/liquidity_input_2000/val_liquidity_segment_1_input_2_1000.csv")
liq_val2 = pd.read_csv("/kaggle/input/findsum-text-data/text/FINDSum-Liquidity/liquidity_input_2000/val_liquidity_segment_2_input_2_1000.csv")
```

```python
from datasets import Dataset, DatasetDict

# Combine dataframes if necessary
liq_test = pd.concat([liq_test0, liq_test1, liq_test2], axis=0)
liq_train = pd.concat([liq_train0, liq_train1, liq_train2], axis=0)
liq_val = pd.concat([liq_val0, liq_val1, liq_val2], axis=0)

# Convert pandas DataFrames to Hugging Face Datasets
test_dataset = Dataset.from_pandas(liq_test)
train_dataset = Dataset.from_pandas(liq_train)
val_dataset = Dataset.from_pandas(liq_val)

# Create a DatasetDict
dataset_dict = DatasetDict({
    'train': train_dataset,
    'validation': val_dataset,
    'test': test_dataset
})
```

## FINDSum-ROO

```python
# import pandas as pd


roo_test0 = pd.read_csv("/kaggle/input/findsum-text-data/text/FINDSum-ROO/roo_input_2000/test_roo_segment_0_input_2_1000.csv")
roo_test1 = pd.read_csv("/kaggle/input/findsum-text-data/text/FINDSum-ROO/roo_input_2000/test_roo_segment_1_input_2_1000.csv")

roo_train0 = pd.read_csv("/kaggle/input/findsum-text-data/text/FINDSum-ROO/roo_input_2000/train_roo_segment_0_input_2_1000.csv")
roo_train1 = pd.read_csv("/kaggle/input/findsum-text-data/text/FINDSum-ROO/roo_input_2000/train_roo_segment_1_input_2_1000.csv")

roo_val0 = pd.read_csv("/kaggle/input/findsum-text-data/text/FINDSum-ROO/roo_input_2000/val_roo_segment_0_input_2_1000.csv")
roo_val1 = pd.read_csv("/kaggle/input/findsum-text-data/text/FINDSum-ROO/roo_input_2000/val_roo_segment_1_input_2_1000.csv")
```

```python
from datasets import Dataset, DatasetDict

# Combine dataframes if necessary
roo_test = pd.concat([roo_test0, roo_test1], axis=0)
roo_train = pd.concat([roo_train0, roo_train1], axis=0)
roo_val = pd.concat([roo_val0, roo_val1], axis=0)

# Convert pandas DataFrames to Hugging Face Datasets
test_dataset = Dataset.from_pandas(roo_test)
train_dataset = Dataset.from_pandas(roo_train)
val_dataset = Dataset.from_pandas(roo_val)

# Create a DatasetDict
dataset_roo = DatasetDict({
    'train': train_dataset,
    'validation': val_dataset,
    'test': test_dataset
})
```

## Combined_ROO_Liquidity_Dataset

```python
# Combine ROO datasets
roo_train = pd.concat([roo_train0, roo_train1], axis=0).sample(frac=1, random_state=42)  # Shuffle
roo_val = pd.concat([roo_val0, roo_val1], axis=0).sample(frac=1, random_state=42)      # Shuffle
roo_test = pd.concat([roo_test0, roo_test1], axis=0).sample(frac=1, random_state=42)   # Shuffle

# Combine Liquidity datasets
liq_train = pd.concat([liq_train0, liq_train1, liq_train2], axis=0).sample(frac=1, random_state=42)  # Shuffle
liq_val = pd.concat([liq_val0, liq_val1, liq_val2], axis=0).sample(frac=1, random_state=42)         # Shuffle
liq_test = pd.concat([liq_test0, liq_test1, liq_test2], axis=0).sample(frac=1, random_state=42)     # Shuffle
```

```python
# Add 'source' column to ROO and Liquidity data
roo_train['source'] = 'ROO'
roo_val['source'] = 'ROO'
roo_test['source'] = 'ROO'

liq_train['source'] = 'Liquidity'
liq_val['source'] = 'Liquidity'
liq_test['source'] = 'Liquidity'
```

```python
# Combine all datasets into single DataFrames
train_combined = pd.concat([roo_train, liq_train], axis=0).sample(frac=1, random_state=42)  # Shuffle combined
val_combined = pd.concat([roo_val, liq_val], axis=0).sample(frac=1, random_state=42)      # Shuffle combined
test_combined = pd.concat([roo_test, liq_test], axis=0).sample(frac=1, random_state=42)   # Shuffle combined
```

```python
from datasets import Dataset, DatasetDict

# Convert combined datasets to Hugging Face Dataset format
train_dataset = Dataset.from_pandas(train_combined)
val_dataset = Dataset.from_pandas(val_combined)
test_dataset = Dataset.from_pandas(test_combined)

# Create a DatasetDict with train, validation, and test splits
final_dataset = DatasetDict({
    'train': train_dataset,
    'validation': val_dataset,
    'test': test_dataset
})
```

```python
final_dataset
```

```python
final_dataset.push_to_hub('ARSynopsis/Combined_ROO_Liquidity_Dataset')
```
